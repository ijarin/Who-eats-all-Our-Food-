```
#bring the USAFAO.csv data as a table
import pandas as pd
import numpy as np
df = pd.read_csv('USAFAO.csv')
df.head(5)
```

```
df_wheat=df.loc[df['Item'] == 'Wheat and products']
df_wheat
```

```
df_wheat1=df_wheat.drop(['Area Code', 'Item Code','Element Code',	'Element',	'Unit',	'latitude',	'longitude'], axis=1)
```

```
df_wheatsum=df_wheat1.groupby(['Area','Item']).sum()      #sum both food and feed
```

```
df_wheatsum
```

```
df_wheatsum.info()  #sum value from year 1962-2013
```

```
Yin=df_wheatsum.values  #take the wheat production values as an array which would be our target class
Yin=Yin.reshape(53,1)
```

```
xin=[1961,1962,1963,1964,1965,1966,1967,1968,1969,1970,1971,1972,1973	,1974,	1975,	1976,	1977,1978,1979,1980,	1981,	1982,	1983,	1984,	1985,	1986,	1987,	1988,	1989,	1990,	1991,	1992,	1993,	1994,	1995,	1996,	1997,	1998,	1999,	2000,	2001,	2002,	2003,	2004,	2005,	2006,	2007,	2008,	2009,	2010,	2011,	2012,	2013]
```

```
Xin=np.array(xin)
Xin.shape
```

```
#Function to create matrix
def calc_matrix(x,m):
  n=len(x)
  #m=3
  A=np.zeros((n,m+1))
  for i in range(m+1):
      A[:,i]=x**i
      #A[:,1]=x
      #A[:,2]=np.square(x)
      #A[:,3]=x**3
  
  return A
```

```
#calculate the matrix when M=2
#calculate the matrix A
Xtrain=calc_matrix(Xin,2)
Xtrain
```

```
X1=np.asarray(range(2014,2020))
Xtest=calc_matrix(X1,2)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model=model.fit(Xtrain, Yin)
predict=model.predict(Xtest)
predict
```
